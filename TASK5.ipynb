```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
legit=df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

**Implementing UnderSampling**

```python
legit_sample=legit.sample(n=104)
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

**Data Splitting**

```python
X=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
print(X)
```

**Data Model Selection**



```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
print(X.shape,X_train.shape,X_test.shape)
```

**Training the Model**

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

**Model Evaluaion**

```python
#For Training Data
train_predict=model.predict(X_train)
train_accuracy=accuracy_score(train_predict,y_train)
print('Accuracy Score of Training Data: ',train_accuracy*100)
```

```python
#For Testing Data
test_predict=model.predict(X_test)
test_accuracy=accuracy_score(test_predict,y_test)
print('Accuracy Score of Testing Data: ',test_accuracy*100)
```
